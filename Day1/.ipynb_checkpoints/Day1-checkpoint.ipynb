```python
#to remove missing values 
print(mel_data.shape)
mel_data = mel_data.dropna(axis=0)
mel_data.shape
```

```python
#selecting prediction target
y = mel_data.Price

#selecting features
mel_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
#selectig specific feature from dataframe
X = mel_data[mel_features]
X.describe()
```

```python
X.head()
```

```python
X.shape
```

```python
#usinf scikit learn to use models
from sklearn.tree import DecisionTreeRegressor
mel_model = DecisionTreeRegressor(random_state=1)
mel_model.fit(X,y)
```

```python
#testing
print("making prediction for the following 5 houses : ")
print(X.head())
print('original prices are')
print(mel_data.Price[:5])
print("the predictions are")
print(mel_model.predict(X.head()))
```

```python
##Model Validation
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = mel_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
## train test splitting
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X,y ,random_state = 0)
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_y)
val_predict = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predict))
```
